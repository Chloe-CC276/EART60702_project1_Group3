```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read data
# spring
data_spring=pd.read_csv("D:/Man/Earth and Environment/spring_data.csv")
original_spring=len(data_spring)

data_spring.head()
```

```python
# summer
data_summer=pd.read_csv("D:/Man/Earth and Environment/summer_data.csv")
original_summer=len(data_summer)

data_summer.head()
```

```python
# autumn
data_autumn=pd.read_csv("D:/Man/Earth and Environment/autumn_data.csv")
original_autumn=len(data_autumn)

data_autumn.head()
```

```python
# winter
data_winter=pd.read_csv("D:/Man/Earth and Environment/winter_data.csv")
original_winter=len(data_winter)

data_winter.head()
```

```python
# filter extreme temperature
# spring
data_spring=data_spring[data_spring['outlier_bool']==True].copy()
spring=data_spring[[
    'Urban Maximum Temperature', 'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]
filtered_spring=len(spring)

print(f"The amount of outlier:{len(spring)}")
spring.head()
```

```python
# summer
data_summer=data_summer[data_summer['outlier_bool']==True].copy()
summer=data_summer[[
    'Urban Maximum Temperature', 'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]
filtered_summer=len(summer)

print(f"The amount of outlier:{len(summer)}")
summer.head()
```

```python
# autumn
data_autumn=data_autumn[data_autumn['outlier_bool']==True].copy()
autumn=data_autumn[[
    'Urban Maximum Temperature', 'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]
filtered_autumn=len(autumn)

print(f"The amount of outlier:{len(autumn)}")
autumn.head()
```

```python
# winter
data_winter=data_winter[data_winter['outlier_bool']==True].copy()
winter=data_winter[[
    'Urban Maximum Temperature', 'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]
filtered_winter=len(winter)

print(f"The amount of outlier:{len(winter)}")
winter.head()
```

```python

data={
    'Season':['Spring','Summer','Autumn','Winter'],
    'Original':[original_spring,original_summer,original_autumn,original_winter],
    'Outliers':[filtered_spring,filtered_summer,filtered_autumn,filtered_winter]
}
df_plot=pd.DataFrame(data)

plt.figure(figsize=(10,7))
plt.bar(df_plot['Season'],df_plot['Original'],label='Original Data')
plt.bar(df_plot['Season'],df_plot['Outliers'],bottom=df_plot['Original'],label='Outliers')

plt.title('Comparison of Total vs Outliers Data',fontsize=18,pad=15)
plt.xlabel('Season',fontsize=14)
plt.ylabel('Number of Days',fontsize=14)
plt.legend()

for i in range(len(df_plot)):
    total=df_plot['Original'][i]
    outlier=df_plot['Outliers'][i]
    percentage=(outlier/total)*100
    plt.text(i,total+(total*0.01),f'{percentage:.1f}%',ha='center',fontweight='bold')

plt.savefig("D:/Man/Earth and Environment/Comparison of Total vs Outliers Data.png",bbox_inches='tight',format='png')
plt.show()
```

```python
# caculate correlation matrix
# spring
corr_spring=spring.corr()
print(corr_spring['Air Temperature'].sort_values(ascending=False))
```

```python
# summer
corr_summer=summer.corr()
print(corr_summer['Air Temperature'].sort_values(ascending=False))
```

```python
# autumn
corr_autumn=autumn.corr()
print(corr_autumn['Air Temperature'].sort_values(ascending=False))
```

```python
# winter
corr_winter=winter.corr()
print(corr_winter['Air Temperature'].sort_values(ascending=False))
```
