```python
import numpy as np 
import pandas as pd 
```

```python
y = pd.read_csv("Original_data.csv")
```

```python
rename_columns = {
    "Time": "Time",
    "TREFMXAV_U": "Urban Maximum Temperature",
    "TREFHT": "Air Temperature",
    "FLNS": "Net Longwave Radiation",
    "FSNS": "Net Solar Radiation",
    "PRECT": "Total Precipitation",
    "PRSN": "Snowfall",
    "QBOT": "Water Vapor Mixing Ratio",
    "UBOT": "Zonal Wind",
    "VBOT": "Meridional Wind",
    "Lat": "Latitude",
    "Lon": "Longitude"
}

# Rename columns
y.rename(columns=rename_columns, inplace=True)
```

```python
y["time"] = pd.to_datetime(y["time"])
```

```python
y["month"] = y["time"].dt.month
```

```python
winter_df = y[y["season"] == "Winter"]
spring_df = y[y["season"] == "Spring"]
summer_df = y[y["season"] == "Summer"]
autumn_df = y[y["season"] == "Autumn"]
```

```python
temp_col = "Air Temperature" 


season_mean = y.groupby("season")[temp_col].transform("mean")
season_std  = y.groupby("season")[temp_col].transform("std")

lower = season_mean - 1.96 * season_std
upper = season_mean + 1.96 * season_std

y["outlier_bool"] = (y[temp_col] < lower) | (y[temp_col] > upper)
```

```python
winter_df.to_csv("winter_data.csv", index=False)
spring_df.to_csv("spring_data.csv", index=False)
summer_df.to_csv("summer_data.csv", index=False)
autumn_df.to_csv("autumn_data.csv", index=False)
```
