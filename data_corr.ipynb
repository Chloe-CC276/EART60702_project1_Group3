```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read data
# spring
data_spring=pd.read_csv("D:/Man/Earth and Environment/spring_data.csv")
data_spring.head()
```

```python
# summer
data_summer=pd.read_csv("D:/Man/Earth and Environment/summer_data.csv")
data_summer.head()
```

```python
# autumn
data_autumn=pd.read_csv("D:/Man/Earth and Environment/autumn_data.csv")
data_autumn.head()
```

```python
# winter
data_winter=pd.read_csv("D:/Man/Earth and Environment/winter_data.csv")
data_winter.head()
```

```python
# filter extreme temperature
# spring
data_spring=data_spring[data_spring['outlier_bool']==True].copy()
spring=data_spring[[
    'Urban Maximum Temperature', 'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]

print(f"The amount of outlier:{len(spring)}")
spring.head()
```

```python
# summer
data_=data_summer[data_summer['outlier_bool']==True].copy()
summer=data_summer[[
    'Urban Maximum Temperature', 'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]

print(f"The amount of outlier:{len(summer)}")
summer.head()
```

```python
# autumn
data_=data_autumn[data_autumn['outlier_bool']==True].copy()
autumn=data_autumn[[
    'Urban Maximum Temperature', 'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]

print(f"The amount of outlier:{len(autumn)}")
autumn.head()
```

```python
# winter
data_=data_winter[data_winter['outlier_bool']==True].copy()
winter=data_winter[[
    'Urban Maximum Temperature', 'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]

print(f"The amount of outlier:{len(winter)}")
winter.head()
```

```python
# caculate correlation matrix
# spring
corr_spring=spring.corr()
print(corr_spring['Air Temperature'].sort_values(ascending=False))
```

```python
# summer
corr_summer=summer.corr()
print(corr_summer['Air Temperature'].sort_values(ascending=False))
```

```python
# autumn
corr_autumn=autumn.corr()
print(corr_autumn['Air Temperature'].sort_values(ascending=False))
```

```python
# winter
corr_winter=winter.corr()
print(corr_winter['Air Temperature'].sort_values(ascending=False))
```

```python
# Urban Maximun Temperature might cause multicollinearity, so exclude
# spring
spring_col=data_spring[[
    'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]
corr_spring_ex=spring_col.corr()
print(corr_spring_ex['Air Temperature'].sort_values(ascending=False))
```

```python
# summer
summer_col=data_summer[[
    'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]
corr_summer_ex=summer_col.corr()
print(corr_summer_ex['Air Temperature'].sort_values(ascending=False))
```

```python
# autumn
autumn_col=data_autumn[[
    'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]
corr_autumn_ex=autumn_col.corr()
print(corr_autumn_ex['Air Temperature'].sort_values(ascending=False))
```

```python
# winter
winter_col=data_winter[[
    'Net Longwave Radiation', 
    'Net Solar Radiation', 'Total Precipitation', 
    'Water Vapor Mixing Ratio', 'Air Temperature', 
    'Zonal Wind', 'Meridional Wind']]
corr_winter_ex=winter_col.corr()
print(corr_winter_ex['Air Temperature'].sort_values(ascending=False))
```

```python
# heat map
# spring
plt.figure(figsize=(12,10))
sns.heatmap(corr_spring_ex,annot=True,cmap='coolwarm',fmt=".2f",linewidths=0.5)
plt.title('Correlation Heatmap of Variables during Extreme Temperature (Spring)',fontsize=20,pad=20)
plt.savefig("D:/Man/Earth and Environment/Correlation Heatmap (Spring).png",bbox_inches='tight',format='png')
plt.show()
```

```python
# summer
plt.figure(figsize=(12,10))
sns.heatmap(corr_summer_ex,annot=True,cmap='coolwarm',fmt=".2f",linewidths=0.5)
plt.title('Correlation Heatmap of Variables during Extreme Temperature (Summer)',fontsize=20,pad=20)
plt.savefig("D:/Man/Earth and Environment/Correlation Heatmap (Summer).png",bbox_inches='tight',format='png')
plt.show()
```

```python
# autumn
plt.figure(figsize=(12,10))
sns.heatmap(corr_autumn_ex,annot=True,cmap='coolwarm',fmt=".2f",linewidths=0.5)
plt.title('Correlation Heatmap of Variables during Extreme Temperature (Autumn)',fontsize=20,pad=20)
plt.savefig("D:/Man/Earth and Environment/Correlation Heatmap (Autumn).png",bbox_inches='tight',format='png')
plt.show()
```

```python
# winter
plt.figure(figsize=(12,10))
sns.heatmap(corr_winter_ex,annot=True,cmap='coolwarm',fmt=".2f",linewidths=0.5)
plt.title('Correlation Heatmap of Variables during Extreme Temperature (Winter)',fontsize=20,pad=20)
plt.savefig("D:/Man/Earth and Environment/Correlation Heatmap (Winter).png",bbox_inches='tight',format='png')
plt.show()
```
